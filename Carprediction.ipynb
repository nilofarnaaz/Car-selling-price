```python
final_dataset=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_dataset.head()
```

```python
#create current year in order to subtract it from year to get how old the car is
final_dataset['Current_Year']= 2020
```

```python
final_dataset.head()
```

```python
#now create Num_of_year--- how old the car is?
final_dataset['Num_of_year']=final_dataset['Current_Year']-final_dataset['Year']
```

```python
final_dataset.head()
```

```python
#now Year and Current_year is of no need as we already create Num_of_year
#So drop it
final_dataset.drop(['Year'],axis=1,inplace=True)
```

```python
final_dataset.head()
```

```python
# drop Current Year
# inplace = True ---->>operation happens in place like permanent operation...if we do not write it may occurs problems that time
final_dataset.drop(['Current_Year'],axis=1,inplace=True)
```

```python
final_dataset.head()
```

```python
# convert categorical features in to one-hot encoded because they are in less num of categories
# get_dummies -->> it is a function under panda which help us to convert all in to one-hot encoded
# drop_first=True -->> First column should also be deleted to actually prevented to dummies variable trap
final_dataset=pd.get_dummies(final_dataset,drop_first=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
#Hyperparameters
import numpy as np
n_estimators =  [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
print(n_estimators)
```

```python
 #Randomized Search CV --->> it finds the best paramters out of these below one

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
#Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,Y_train)
```

```python
predictions=rf_random.predict(X_test)
```
